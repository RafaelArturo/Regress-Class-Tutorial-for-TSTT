**Machine Learning** is a tool used in AI that gives the computers the ability to learn without being explicitly programmed to do so. It can be classified under three (3) categories or procedures:

1) **Supervised Learning** - In supervised learning, we are aware of the values or labels of the outputs for given inputs from the previous data that we have. The machine learns the input to output mappings based on these past records to predict future outputs of new data. One use case example is an email spam filter which detects, given an input of emails, what is spam and what is not spam. Previous data would be already classified emails as spam or not spam. Examples of supervised learning procedures include linear regression, multiple regression and logistic regression.

2) **Unsupervised Learning** - In unsupervised learning, we are not aware of the values or labels of the outputs for given inputs from the previous data that we have. We use the data to infer clusters or groups of the data-points by similar properties that are shared in the dataset and in some cases detect anomalies. Examples of unsupervised learning procedures include K-means clustering, hierarchical clustering, db scan clustering and anomaly detection.

3) **Reinforcement Learning** - reinforcement learning is semi-supervised learning in that some part of the data is labelled, and some part is not labelled for future actions. The model learns slowly by seeing past data and will learn and then apply its knowledge to the new data as it comes up. These procedures are applied to systems that perform games like chess and backgammon, and even in robotics. The machine will know what to do next but there is uncertainty about the future until the previous task is performed.

<br>
Supervised learning procedures are the most commonly used ones.

### 2. Regression Analysis

According to Wikipedia, "regression analysis is a set of statistical processes for estimating the relationships between a dependent variable (often called the 'outcome variable') and one or more independent variables (often called 'predictors', 'covariates', or 'features')."

For regression analysis, some use the term, **general linear models**, in the broad scope of things, if there are multiple dependent variables. These variables are correlated (generally a change in one of these affects the others) and hence, multivariate regression analysis would be performed.

In this section, we will focus primarily on linear regression where there is only one (1) dependent variable.

In the classification section, we will discuss the broader sense of **generalized linear models** with which general linear models fall under. The terminology may sound confusing now but hopefully by the end of the session we should be able to have a better sense of things.

<br>
<br>
<br>

#### Linear Regression:

Linear Regression: In a nutshell, we look to take input value(s) and map them on the set of real-numbers. The output is a continuous variable, which is what we want to predict given the input value(s). The hard part is knowing how to model the data (which variables to keep, what type of relationships to use, etc.) and that is why different models are chosen. Examples include predicting stock prices, predicting market value of houses, etc.

Linear regression falls into the category of general linear models with which there is only one dependent variable. The dependent variable is modelled based on none, one or more than one independent variables. In addition, the dependent variable is continuous, while each independent variable can be either continuous or categorical.

<br>

As a result, in linear regression, we generally condition the dependent variable, $Y$, on the independent variable(s), $X$, in the form of $Y|X$.
It should be noted however, that it can be common to just write $Y$ but the understanding is that we are conditioning $Y$ on $X$.

<br>
<br>

**Linear Regression Model**

A linear regression model has the form:

\begin{equation}
    Y = β_{0} + β_{1} X_{1} + \ldots + β_{p} X_{p} + ϵ
    \label{eq:linearregmodel} \tag{1}
\end{equation}

where $Y$ is the dependent variable
<br>
$X_{1}$, $X_{2}$, $\ldots$, $X_{p}$ are the p independent variables
<br>
$β_{0}$, $β_{1}$, $\ldots$, $β_{p}$ are unknown coefficient parameters
<br>
and $ϵ$ is the random error or true residual which is normally distributed with mean, 0, and variance, $σ^{2}$.

<br>
<br>

**Answer 5:**
<br>
$β_{0}$ gives us the expected value of $Y$ (or $Y|X$) either when the all the $X_{j}$'s are 0, all the $β_{j}$'s are 0 or sum of all the $β_{j}X_{j}$'s for j = 1, $\ldots$, p are 0.

The definition of $β_{j}$ may vary depending on whether $X_{j}$ is a continuous variable or categorical variable.

**When we have Continuous Independent Variables**

When $X_{j}$ is continuous, $β_{j}$ for j = 1, $\ldots$, p tells us that a one unit increase in $X_{j}$, holding all other units constant, would result in an expected increase in $Y$ (or $Y|X$) by $β_{j}$ units.


<br>
<br>


If we take the expectation of model $\eqreg{eq:linearregmodel} above, we get:

\begin{equation}
    E(Y|X) = β_{0} + β_{1} X_{1} + \ldots + β_{p} X_{p}
    \label{eq:expectlinearregmodel} \tag{2}
\end{equation}

<br>
<br>

Let's use an example of equation $\eqref{eq:linearregmodel}$ above with p = 2 and let $β_{0}$ = 0.45, $β_{1}$ = 2.37 and $β_{2}$ = 5.3. Now, we will have the following model:

\begin{equation}
    Y = 0.45 + 2.37 X_{1} + 5.3 X_{2} + ϵ  
\end{equation}

<br>

The expectation would result in:

\begin{equation}
    E(Y|X) = 0.45 + 2.37 X_{1} + 5.3 X_{2}
\end{equation}

Thus,
<br>
$β_{0}$ = 0.45 gives us the expected value of $Y$ (or $Y|X$) when all $X_{j}$'s are 0.

\begin{equation}
    E(Y|β_{1}X_{1}+β_{2}X_{2}=0) = 0.45
\end{equation}

<br>

$β_{1}$ = 2.37 tells us that a one unit increase in $X_{1}$, holding all other units constant, would result in an expected increase in $Y$ (or $Y|X$) by 2.37 units.

\begin{equation}
    E(Y|X_{1}+1, X_{2}) - E(Y|X_{1}, X_{2}) = 2.37
\end{equation}

<br>

$β_{2}$ = 5.3 tells us that a one unit increase in $X_{2}$, holding all other units constant, would result in an expected increase in $Y$ (or $Y|X$) by 5.3 units. 

\begin{equation}
    E(Y|X_{1}, X_{2}+1) - E(Y|X_{1}, X_{2}) = 5.3
\end{equation}

<br>
<br>
<br>

**When we have Categorical Independent Variables**

When we have independent categorical variables, we can assign numerical values to it. We assign dummy variables in these cases which can generally only take two values, 0 or 1. For instance, male=0/female=1 or no=0/yes=1, etc. The dummy variable acts like an on and off switch for an option. Usually, when one switch is on, the other switches are off. 

<br>

If we were to cater for a categorical variable with more than two options, say the colour options: platinum, gold, silver or bronze, which has 4 options, multiple dummy variables would be assigned to represent these scenarios in the model. A general rule of thumb for i options in a categorical variable is to use i-1 dummy variables if the intercept variable is available. Luckily, in programming these models, some software packages and code can do this for you automically, so you can generally just assign the names for each option in a vector.

For the colour example, suppose we have a dependent variable Y and the colour options are the only other variable in question. Then, by having 4 options, we will need 3 variables apart from the intercept variable. Each of these 4 variables would represent a colour option, and the other variables would be set to 0. So,

\begin{equation}
    Y = β_{0} + β_{1} X_{1} + β_{2} X_{2} + β_{3} X_{3} + ϵ
\end{equation}
and
\begin{equation}
    E(Y|X) = β_{0} + β_{1} X_{1} + β_{2} X_{2} + β_{3} X_{3}
\end{equation}

where $β_{0}$ represents the expected value of $Y$ when we choose 1 colour, say platinum, provided that $X_{1}$, $X_{2}$ and $X_{3}$ are set to 0.
$β_{0} + β_{1}$ represents the expected value of $Y$ when we choose another colour, say gold, provided that $X_{1}$ is set to 1, and $X_{2}$ and $X_{3}$ are set to 0.
$β_{0} + β_{2}$ represents the expected value of $Y$ when we choose another colour, say silver, provided that $X_{2}$ is set to 1, and $X_{1}$ and $X_{3}$ are set to 0.
$β_{0} + β_{3}$ represents the expected value of $Y$ when we choose another colour, say bronze, provided that $X_{3}$ is set to 1, and $X_{1}$ and $X_{2}$ are set to 0.

Notice that we always use the $β_{0}$ value for each colour option. As a result, the value of the other $β$'s depend on this value. In general, the $β_{j}$'s for j = 1, $\ldots$, p represent the expected change in value of $Y$ when we choose the j$^{th}$ additional colour opposed to the base colour (associated with $β_{0}$), holding all other variables constant.

**Note:** We can have models with both continuous and categorical independent variables.

**Applying Obervations to the Model**

The data to fit the model would consist of n independent observations of $Y$ with corresponding $X_{j}$'s. We will now have the following model:

\begin{equation}
    Y_{i} = β_{0} + β_{1} x_{i1} + \ldots + β_{p} x_{ip} + ϵ_{i}
    \label{eq:obslinearregmodel1} \tag{3}
\end{equation}

where
<br>
i = 1, $\ldots$, n
<br>
$Y_{i}$ is the dependent variable for the i$^{th}$ observation
<br>
$x_{i1}$, $x_{i2}$, $\ldots$, $x_{ip}$ are the p independent variables for the i$^{th}$ observation
<br>
$β_{0}$, $β_{1}$, $\ldots$, $β_{p}$ are unknown parameters
<br>
and $ϵ_{i}$ is the i$^{th}$ error term


<br>
<br>

**Assumptions of the Linear Regression Model**

The assumptions of the linear regression model are:
<br>
$ϵ_{i}$'s are independent and have identically Normal distributions with mean, μ = 0 and constant variance, $σ^{2}$
<br>
i.e. $ϵ_{i}$'s ~ i.i.d. N(0, $σ^{2}$)
<br>
across all observations in the population.

<br>
From these assumptions:
<br>

- The observations should be independently conducted. We test independence of observations by plotting residuals, $\hat{ϵ}_{i}$ vs time, i (order, i, in which the n variables were obtained).
<br>

- A fancy term used in place of constant variance is *homoscedasticity*. This just means that the variance, $σ^{2}$, is the same across all observations. We test constant variance by plotting residuals, $\hat{ϵ}_{i}$, vs fitted values, $\hat{Y}_{i}$ and determine if there is a constant horizontal band of points centred around 0. This plot can also detect cases of non-linearity in the model (at least one independent variable has a non-linear relationship with the dependent variable), but LOESS (or LOWESS) curves in this plot are probably better to include as well. When the homoscedasticity assumption is violated, we end up getting heteroscedasticity (non-constant variance across observations). Heteroscedasticity is troublesome because:
    1) some observations are less reliable than others and should be down-weighted in the fitting procedure.
    2) it is a sign that ordinary least-squares estimation (OLS) is not optimal and we can do better by switching to weighted least-squares (WLS).
    3) it can distort standard errors (deviations) of the estimates.
<br>

- Moderate departures from normality may not significantly impact linear regression to work effectively, especially if the sample size is large. We test normality using normal q-q plots (sample quantile vs theoretical quantile) or hypothesis tests like the Shapiro–Wilks test. For the normal q-q plots, we should get a diagonal alignment of points. “Rather than testing the null hypothesis of normality-which is false anyway-we should focus on whether the departures from normality exhibited by our data are serious enough to adversely impact the procedures that we want to use.” *- STAT 6120 Lecture 2 Notes*

\begin{equation}

    \label{eq:obslinearregmodel2} \tag{4}
\end{equation}


where we can assign the value of 1 for $x_{i1}$ for all i.

<br>

<font color='red'>

**Note:** If we convert $\eqref{eq:obslinearregmodel1}$ to $\eqref{eq:obslinearregmodel2}$ with a defined value for p in $\eqref{eq:obslinearregmodel1}$, say p = 5, it will be p+1 in $\eqref{eq:obslinearregmodel2}$, i.e. p+1 = 6.

<br>
<font color='black'>

\end{equation}

One of the main goals in regression is to find the value of $β$ that makes

\begin{equation}
    Xβ = β_{1} X_{1} + β_{2} X_{2} + \ldots + β_{p} X_{p}
\end{equation}

close to Y.

<br>

One approach is to use *the method of ordinary least-squares*. For this method, we need to find the $\hat{β}$ for which 

\begin{equation}
    \hat{Y} = X\hat{β} = \hat{β}_{1} X_{1} + \hat{β}_{2} X_{2} + \ldots + \hat{β}_{p} X_{p}
\end{equation}

will minimise the Euclidean distance between $Y$ and $\hat{Y}$; i.e. minimise ||$\hat{ϵ}$||, where $\hat{ϵ}$ = $Y$ - $\hat{Y}$. We are essentially finding the $\hat{β}$ for which $\hat{Y}$ is the projection of $Y$ on the linear space spanned by the columns of $X$ (i.e. $X_{1}, \ldots, X_{p}$).

<br>

In the proof, we use ||$\hat{ϵ}$||$^2$ = ($Y$ - $\hat{Y}$)$^{T}$($Y$ - $\hat{Y}$), and it can be shown that

\begin{equation}
    \hat{Y} = X(X^{T}X)^{-1}X^{T}Y
\end{equation}
and hence, $\hat{Y}$ = $X\hat{β}$ gives us,
\begin{equation}
    \hat{β} = \begin{bmatrix}
    \hat{β}_{1} & \hat{β}_{2} & \cdots & \hat{β}_{p}
    \end{bmatrix}
    ^{T} = (X^{T}X)^{-1}X^{T}Y
    \label{eq:betahat} \tag{9}    
\end{equation}

||$\hat{ϵ}$||$^2$ is called the residual sum of squares or sum of squared residuals or error sum of squares.

From this and $Y|X$ ~ N($Xβ$, $σ^{2}I$), we can get:
<br>
<center> $\hat{β}|X$ ~ N($β$, $σ^{2}(X^{T}X)^{-1}$) </center>

<br>
<br>

**Some Linear Regression Models**

1) **No-Predictors Model:** *We have an intercept term and no independent variables.*

\begin{equation}
    Y_{i} = β_{0} + ϵ_{i}
    \label{eq:nopredlinearregmodel} \tag{10}
\end{equation}

<br>

<font color='red'>

**Note:** It can be shown for the no-predictors model that:

\begin{equation}
    \hat{β}_{0} = \bar{Y} = \sum_{i=1}^{n}Y_{i}
\end{equation}


<font color='black'>

<br>


2) **Simple Linear Regression Model:** *We have an intercept term and one independent variable.*

\begin{equation}
    Y_{i} = β_{0} + β_{1} x_{i1} + ϵ_{i}
    \label{eq:simplinearregmodel} \tag{11}
\end{equation}


3) **Multiple Linear Regression Model:** *We have an intercept term and multiple independent variables.*

\begin{equation}
    Y_{i} = β_{0} + β_{1} x_{i1} + \ldots + β_{p} x_{ip} + ϵ_{i}
    \label{eq:multlinearregmodel} \tag{12}
\end{equation}

<br>


4) **Polynomial Regression Model:** *We have an intercept term and multiple independent variables, some of which could be of a higher degree polynomial.*

\begin{equation}
    Y_{i} = β_{0} + β_{1} x_{i1} + β_{2} x_{i1}^2 + \ldots + β_{p} x_{i1}^p + ϵ_{i}
    \label{eq:polyregmodel} \tag{13}
\end{equation}

This is still a linear function of the $β$'s because you can treat $x_{i1}^j$, just like other variables, $x_{ij}$, for j = 1, $\ldots$, p in the multiple linear regression model.

<br>

<font color='red'>

**Note:** It is possible to have other independent variables and even interactions between independent variables in the model.

<br>

<font color='black'>

So, you can use the same principles for linear regression for a model like this:

\begin{equation}
    Y_{i} = β_{0} + β_{1} x_{i1} + β_{2} x_{i1}^{5} + β_{3} x_{i2} + β_{4} x_{i1} x_{i2} + ϵ_{i}
    \label{eq:mixedregmodel} \tag{14}
\end{equation}

<br>

<font color='red'>

**Note:** It is even possible to have various transformations of the dependent and/or independent variables in the model. We can use these transformations so that the assumptions of the linear regression are met. The interpretation of the coefficients can, however, get confusing.

<br>

<font color='black'>

So, you can use the same principles for linear regression for a model like this:

\begin{equation}
    ln(Y_{i}) = β_{0} + β_{1} log(x_{i1}) + β_{2} x_{i1}^{2} + β_{3} \frac{1}{x_{i2}} + ϵ_{i}
    \label{eq:transformedregmodel} \tag{15}
\end{equation}

<br>

<font color='red'>

**Note:** It is possible to assume no intercept term ($β_{0}$) in the model.

$MORPH = \frac{WAIST}{HIPS}$

and

\begin{equation}
    E(CHOL|BMI, MORPH) = β_{0} + β_{1} BMI + β_{2} MORPH

```python
# Importing the body.txt file.

# Body is a .txt file with a single space delimiter.
body = pd.read_csv('Datasets for Regression & Classification/body.txt', sep="\s+", header='infer')

print("The dimensions of the body.txt data frame is", body.shape, "\n\n")      # Look at the dimensions of the data frame.

print("The body.txt data frame looks like:\n")
body     # Look at the rows of data.

#?pd.read_csv
```

Next, we transform the data to get *BMI* and *MORPH*.

```python
# We have to create two more variables, BMI and MORPH
# using some variables in the dataset.


# First, let's look at BMI:
# BMI = WEIGHT (in kgs) / (HEIGHT (in m))^2

# Notice the change in the units for WEIGHT and HEIGHT.
# So, let's convert them:

# Weight from pounds to kgs
# 1 pound = 0.45359237 kgs
WEIGHT_kg = 0.45359237 * body[['weight']]

# Height from inches to m
# 1 inch = 0.0254 m
HEIGHT_m = 0.0254 * body[['height']]

BMI = np.divide(WEIGHT_kg, HEIGHT_m ** 2)    # ** represents ^ in HEIGHT^2


# Next, let's look at MORPH:
# MORPH = WAIST / HIPS
MORPH = np.divide(body[['waist']], body[['hips']])
```

**Evaluating the Summary of the Multiple Linear Regression for the body dataset:**

```python
Xbody = np.c_[BMI, MORPH]        # Concatenating BMI and MORPH.
Xbody = sm.add_constant(Xbody)   # Adding a column of 1's to the X's.

# Converting X into 1 dataframe. This is really for the appearance in the outputs.
Xbody = pd.DataFrame(Xbody, columns=['Intercept', 'BMI', 'MORPH'])

print("The X matrix looks like:\n", Xbody)

Ybody = body[['chol']]
print("\nThe Y vector looks like:\n", Ybody)

# This code acts similar to the 'lm' function in R to generate the OLS fit among other things.
lin_reg = sm.OLS(Ybody, Xbody).fit()

# This provides a great summary for linear regression.
lin_reg.summary()
```

There are a lot of results outputted in the summary.

For now, we can look at the section with the coefficients.

1) The OLS values of the coefficients for the *Intercept*, *BMI* and *MORPH* (i.e. $\hat{β}_{0}$, $\hat{β}_{1}$ and $\hat{β}_{2}$) are found in the **coef** column.

2) The **std err** is the standard error of the coefficients (i.e. $SE(\hat{β}_{j}) = \sqrt{S^{2}(X^{T}X)_{jj}^{-1}}$ ; where $S^{2}$ = $\frac{|\hat{ϵ}|}{n-p}$ = $\frac{1}{n-p}\sum _{i=1}^{n}(Y_{i} - \hat{Y}_{i})^2$ is the Mean Squared Error).

3) The values in **t** are the test statistics for each coefficient. The t-statistic is found by dividing the value in the coef column with the value in the std err column (i.e. t = $\frac{\hat{β}_{j}}{SE(\hat{β}_{j})}$). This is used in hypotheses for testing $H_{0}$: $β_{j}$ = 0 vs $H_{1}$: $β_{j}$ ≠ 0.

4) The values in **P>|t|** are the corresponding p-values for the hypotheses tests. For the $α$% level of significance, $H_{0}$ is rejected if the p-value is less than or equal to $\frac{α}{100}$.

5) The columns under **[0.025	0.975]** indicate the boundaries of 95% confidence interval (CI) for each coefficient. The CIs can be used for the hypotheses tests as well. $H_{0}$ is rejected at the 5% significance level if 0 lies in the CI.

The output also contains performance measures such as **R-squared** ($R^{2}$), **Adj. R-squared** ($R^{2}_{adj}$), **Log-Likelihood**, Akaike Information Criterion (**AIC**) and Bayesian Information Criterion (**BIC**).

$R^{2}$ is the squared sample correlation between $Y$ and $\hat{Y}$. This is also known as the **coefficient of determination**. $R^{2}$ gives us an idea as to the proportion of the variation in the dependent variable ($Y$) explained by the independent variables ($x$'s other than column of 1's). As we add more variables to the model, $R^{2}$ cannot get worse, but these additional variables may be meaningless to add. $R^{2}_{adj}$ attempts to address this.

The **F-statistic** and **Prob(F-statistic)** are the F-statistic and corresponding p-value for the omnibus F-test which tests $H_{0}$: $β_{1}$ = $\ldots$ = $β_{p}$ = 0 vs $H_{1}$: At least one $β_{j}$ ≠ 0, for j = 1, $\ldots$, p; (i.e. whether adding independent variables makes the model significantly better than using the no-predictors model.).

The **Omnibus** value and **Prob(Omnibus)** are not to be confused with the that from the omnibus F-test. These more have to do with the skewness and kurtosis of the residuals and are tests for normality of the residuals.

**Checking the Assumptions:**

**Plot of Residuals vs Fitted Values**

```python
# Function for plotting the Residuals vs Fitted Values:
def linearity_test(model, y):
    '''
    Function for visually inspecting the assumptions of constant variance and linearity in a linear regression model.
    It plots Residuals vs Fitted Values with a LOESS curve.
    
    Args:
    * model - fitted OLS model from statsmodels
    * y - observed values
    '''
    fitted_vals = model.predict()
    resids = model.resid

    sns.regplot(x=fitted_vals, y=resids, lowess=True, line_kws={'color': 'red'})
    plt.title('Residuals vs. Fitted Values')
    plt.xlabel('Fitted Values')
    plt.ylabel('Residuals')

# Running the function.
linearity_test(lin_reg, Ybody)
```

**Normal Q-Q Plot and Hypotheses Tests for Normality**

```python
from scipy import stats

# Function for plotting the Normal Q-Q plot:
def normality_of_residuals_test(model):
    '''
    Function for drawing the normal QQ-plot of the residuals and running 4 statistical tests to 
    investigate the normality of residuals.
    
    Arg:
    * model - fitted OLS models from statsmodels
    '''
    sm.ProbPlot(model.resid).qqplot(line='s');
    plt.title('Q-Q plot');

    jb = stats.jarque_bera(model.resid)
    sw = stats.shapiro(model.resid)
    ks = stats.kstest(model.resid, 'norm')
    ad = stats.anderson(model.resid, dist='norm')
    
    print(f'Jarque-Bera test ---- statistic: {jb[0]:.4f}, p-value: {jb[1]}')
    print(f'Shapiro-Wilk test ---- statistic: {sw[0]:.4f}, p-value: {sw[1]:.4f}')
    print(f'Kolmogorov-Smirnov test ---- statistic: {ks.statistic:.4f}, p-value: {ks.pvalue:.4f}')
    print(f'Anderson-Darling test ---- statistic: {ad.statistic:.4f}, 5% critical value: {ad.critical_values[2]:.4f}')
    print('If the returned AD statistic is larger than the critical value, then for the 5% significance level, the null hypothesis that the data come from the Normal distribution should be rejected. ')
    
normality_of_residuals_test(lin_reg)
```

### 3. Generalized Linear Models

“Prior to the 1980's, when linear regression was the only tool available, data analysts were taught to transform variables to make them fit into the framework that they knew. They were taught to change the data to accommodate the model.” *- STAT 6120 Lecture 2 Notes*

They would use transformation methods like power transformations and variance-stabilising transformations to get the model to follow the assumptions of normality and homoscedasticity (constant variance) respectively by reducing skewness and heteroscedasticity respectively, and then they would apply the new parameters to normal linear regression models.

“Nowadays, we have a greater variety of models and software available, so we can now change the model to fit the data.” *- STAT 6120 Lecture 2 Notes*

Specifically, we can use generalized linear models (GLMs) to keep to the original scale.

For example, there could be the situation “where the response is a frequency or count, i.e. a number of occurrences in a span of time or region of space. If Y is a count, it is natural to suppose that it may arise from a Poisson process, and therefore it would have a Poisson distribution. If Y is distributed as Poisson with mean, μ, then Var(Y) = μ, so the variance is proportional to the mean. The variance-stabilising transformation for a Poisson count is the square root. In the old days, people used to analyse count data by linear regression on the square-root scale. Nowadays, analysts are much more likely to apply techniques of loglinear modelling, which are based on the Poisson likelihood.” *- STAT 6120 Lecture 2 Notes*

Multiple Linear (Normal) Regression Model Approach:
\begin{equation}
    \sqrt{Y} = β_{0} + β_{1} X_{1} + \ldots + β_{p} X_{p} + ϵ
\end{equation}

GLM Approach - Loglinear Model:
\begin{equation}
    log[E(Y|X)] = log(μ) = β_{0} + β_{1} X_{1} + \ldots + β_{p} X_{p}
\end{equation}

<br>
<br>
<br>

A GLM consists of three (3) components:
<br>

1) **The Random Component** - The random component is based on the identification of the dependent variable (Y) and its probability distribution.
<br>

2) **The Systematic Component** - The systematic component identifies the set of explanatory variables ($X_{1}$, $\ldots$, $X_{p}$).
<br>

3) **The Link Function** - The link function, identifies a function of the mean that is a linear function of the explanatory variables.

\begin{equation}
    g[E(Y|X)] = g(μ) = β_{0} + β_{1} X_{1} + \ldots + β_{p} X_{p}
    \label{eq:glmlinkfunct} \tag{16}
\end{equation}

<br>

1) **Some Distributions of Random Components**

<br>

- *Multiple Linear (Normal) Regression* - The random component, $ϵ$, for the multiple linear regression model has a **normal distribution** with a mean of 0 and constant variance, $σ^{2}$.
<br>

- *Gamma Regression* - If the dependent variable is continuous with a skewed distribution (not symmetric like the normal distribution) and the variation increases with the mean, the random component, $ϵ$, can have a **gamma distribution**.
<br>

- *Binary Logistic Regression* - For binary logistic regression, we only have two (2) possible outcomes, 0 or 1. The random component, $ϵ$, for the binary logistic regression model has a **binomial distribution**.
<br>

- *Poisson Regression* - For poisson regression, we have count data which measures the number of events in a fixed area and/or length of time and the mean of the response equals its variance, $E(Y) = Var(Y)$. The random component, $ϵ$, for the poisson regression model has a **poisson distribution**.
<br>

- *Negative Binomial Regression* - For negative binomial regression, we have count data where $E(Y) < Var(Y)$ (overdispersion). (The negative binomial distribution seeks to measure the time or number of trials until a certain number of successes, say k, occurs.) The random component, $ϵ$, for the negative binomial regression model has a **negative binomial distribution**.

<br>

2) **The Systematic Component**

- The independent variables can consist of continuous and/or categorical variables.

<br>

3) **Some Common Link Functions**

- *Multiple Linear (Normal) Regression* - The multiple linear regression model has an **identity link function** of the form:

\begin{equation}
    g(μ) = μ
    \label{eq:identlinkfunct} \tag{17}
\end{equation}
<br>

- *Gamma Regression* - The gamma regression model can have an identity link function or a **negative inverse link function** of the form:

\begin{equation}
    g(μ) = -\frac{1}{μ}
    \label{eq:neginvlinkfunct} \tag{18}
\end{equation}
<br>

- *Binary Logistic Regression* - The binary logistic regression model has a **logit link function** of the form:

\begin{equation}
    g(μ) = ln\left(\frac{μ}{1-μ} \right)
    \label{eq:logitlinkfunct} \tag{19}
\end{equation}
<br>

- *Poisson Regression* and *Negative Binomial Regression* - Each of these models can have an identity link function but it is typical to use a **log link function** of the form:

\begin{equation}
    g(μ) = ln(μ)
    \label{eq:loglinkfunct} \tag{20}
\end{equation}

<br>

<font color='red'>
    
**Note:** We use $ln$ to represent the natural logarithm. It is common to see the use of $log$ in place of $ln$, but this $log$ is meant to represent the natural logarithm and not base 10 logarithm as well.

<br>
<br>

<font color='black'>

**Some Assumptions of GLMs**

Some assumptions of the generalized linear model are:
<br>

- The observed data, $Y_{1}$, $Y_{2}$, ..., $Y_{n}$, are independently distributed; hence, the cases are independent.
<br>

- The distribution of $Y_{i}$ typically assumes a distribution from an exponential family (e.g. normal, binomial, poisson, multinomial, $\dots$). This distribution of the dependent variable follows the same distribution as that from the random component, $ϵ_{i}$.
<br>

- A linear relationship between the dependent variable and the independent variables is not necessarily assumed; rather, a linear relationship between the link function of the response variable and the explanatory variables is assumed, and the linearity is based on the $β$'s. As a result, the independent (explanatory) variables can even be the power terms or some other nonlinear transformations of the original independent variables.
<br>

- The homogeneity of variance does not need to be satisfied. In fact, it may not be even possible in many cases given the model structure.
<br>

- Errors need to be independent but not necessarily normally distributed.
<br>

- Maximum likelihood estimation (MLE) or Bayesian techniques rather than the method of ordinary least-squares (OLS) would be used to estimate the parameters for certain models.
<br>

- There may be additional assumptions depending on the model employed.

### 4. Classification Analysis

According to Wikipedia, "classification is the problem of identifying to which of a set of categories (sub-populations) a new observation belongs, on the basis of a training set of data containing observations (or instances) whose category membership is known." It is a supervised learning procedure. The unsupervised version of this is called clustering.

The main difference between regression and classification is based on the dependent variable. For regression, the dependent variable can take on a real-valued number, while for classification, the dependent variable assumes categorical values or represents categories or classes.

Each independent variable can be either continuous or categorical as with regression.

An example of classification is the email spam filter, which filters emails as "spam" or "not spam". This is considered binary classification, in which there are only two (2) classes that the dependent variable can assume. For multiclass classification, there are more than two (2) classes that the dependent variable can assume.

There are various algorithms for classification. Some of these include:
- Linear classifiers
 - Fisher's linear discriminant
 - Binary logistic regression
 - Multinomial logistic regression
 - Ordinal logistic regression
 - Naive Bayes classifier
- Support vector machines
 - Least squares support vector machines
- Quadratic classifiers
- Decision trees
 - Random forests
- Neural networks


Linear classifiers classify the dependent variables based on the linear combination of the features. We use a linear prediction function (similar to E(Y|X) in regression) which has the general form:

\begin{equation}
    score(x_{i}, k) = x_{i}^{T} β_{k}
    \label{eq:linpredfunct} \tag{20}
\end{equation}

where i represents the i$^{th}$ observation
k represents the k$^{th}$ class
$x_{i}$ is the feature vector or vector of independent variables for the i$^{th}$ observation
$β_{k}$ is the vector of coefficients corresponding to the k$^{th}$ class

The class with the highest score will be the one that is estimated for the dependent variable.

Suppose we have a classification algorithm which identifies who a person will support (Independent, PNM, UNC, None) in the national elections based on certain features (location, age, income, etc.). If for the m$^{th}$ individual, the highest score is under the class of "Independent", then that is who we assign for the m$^{th}$ dependent variable.

<br>
<br>

**Sources:**
- Wikipedia (https://en.wikipedia.org/wiki/Statistical_classification)

In binary classifiers, we may have classes 0 and 1, and if the score is based on a probability, we can just use the value of one of the scores and apply a function to it to class. For instance, we may have
\begin{equation}
    Y_{i} = \left\{\begin{matrix}
    0, & if g(x_{i}^{T} β) > threshold\\ 
    1, & otherwise
    \end{matrix}\right.
\end{equation}

where $g$ is the function that converts $x_{i}^{T} β$ into a probability (such as $P(Y_{i} = 1| x_{i}, β)$) and the probability above a certain threshold would be part of a class.

**Logistic Regression**

Binary logistic regression was mentioned in Generalized Linear Model. There are various types of logisitic regression depending on the situation at hand. Some of these are:
<br>

- Binary Logistic Regression: Binary logisitic regression is used when the response variable has two (2) possible outcomes.
<br>

- Nominal (Multinomial) Logistic Regression: Nominal logistic regression is used when the response variable has more than two (2) possible categories with no natural ordering to the categories. Some examples of nominal dependent variables are colour (such as red, white, black, blue) or departments at a business (such as human resource, management, analytics).
<br>

- Ordinal Logistic Regression: Ordinal logistic regression is used when the response variable has more than two (2) possible categories with natural ordering to the categories, but the ranking of the categories do not necessarily mean the intervals between them are equal. Some examples of ordinal dependent variables are likely to recommend scores (such as those on a 0-10 scale) or levels of highest education certificate (such as primary, secondary, tertiary).
<br>

We will just focus on binary logistic regression. We use probabilities and odds for logistic regression. "Logistic regression is applicable, for example, if:
<br>

- we want to model the probabilities of a response variable as a function of some explanatory variables, e.g. "success" of admission as a function of gender.
<br>

- we want to perform descriptive discriminate analyses such as describing the differences between individuals in separate groups as a function of explanatory variables, e.g. student admitted and rejected as a function of gender
<br>

- we want to predict probabilities that individuals fall into two categories of the binary response as a function of some explanatory variables, e.g. what is the probability that a student is admitted given she is a female
<br>

- we want to classify individuals into two categories based on explanatory variables, e.g. classify new students into "admitted" or "rejected" group depending on their gender." *- STAT 6160 Notes*

Recall from the Statistics & Probability Session that the probability that an event, $E$, occurs is written as $P(E)$ and the probability that $E$ does not occur is $P(\bar{E})$. Also, $0 ≤ P(E) ≤ 1$. $P(E)=0$ means that $E$ surely does not occur, and if $P(E)=1$, then $E$ surely occurs. The closer $P(E)$ is to 1 means the greater chance of that event occurring.

The **odds** of $E$ is the ratio of the probability of $E$ occurring and the probability of $E$ not occurring.

<br>

<center> $Odds = \frac{P(E)}{1-P(E)} = \frac{P(E)}{P(\bar{E})}$ <center\>
<br>

In addition, $0 < Odds < \infty$ and $-\infty < ln(Odds) < \infty$.


Essentially, what we trying to do is convert this transformation concept (from $P(E)$ to $ln(Odds)$) to the dependent variable, $Y$. Consider $Y_{i}=0$ as the event that $Y$ does not occur for the i$^{th}$ observation, and $Y_{i}=1$ as the event that $Y$ occurs for the i$^{th}$ observation.

The range of $P(Y_{i}=1)$ values is in the interval $[0, 1]$ (i.e. $0 ≤ P(Y_{i}=1) ≤ 1$) to get the range of $ln(Odds)$ in the interval, $[-\infty, \infty]$ the space of a normal distribution using the GLM components.

Using equations (16) and (19), we get that for a binary logistic regression, we want:

\begin{equation}
    logit(μ_{i}) = ln\left(\frac{μ_{i}}{1-μ_{i}} \right) = β_{0} + β_{1} x_{i1} + \ldots + β_{p} x_{ip} = Xβ
    \label{eq:logisticreg} \tag{21}
\end{equation}

and $μ_{i}$ ~ $Binomial(1, \pi_{i}) \equiv Bernoulli(\pi_{i})$

But what is $μ_{i}$? $μ_{i}$ would represent the probability that the event occurs because for a $Bernoulli(\pi)$, $E(Y_{i} = 1| X) = 0 \cdot P(Y_{i} = 0| X) + 1 \cdot P(Y_{i} = 1| X) = P(Y_{i} = 1| X)$.

```python
bodysizelogitmod.summary2() # Another summary option.
```

<br>
<br>

The logistic regression model can be written as:


\begin{equation}
    logit(\pi_{i}) = ln\left(\frac{\pi_{i}}{1-\pi_{i}} \right) = β_{0} + β_{1}x_{i}
\end{equation}

which is estimated with

\begin{equation}
    logit(\hat{\pi}_{i}) = ln\left(\frac{\hat{\pi}_{i}}{1-\hat{\pi}_{i}} \right) = \hat{β}_{0} + \hat{β}_{1}x_{i} = -21.9924 + 0.7404 \cdot x_{i}
\end{equation}

<font color='red'>
    
**Note:** The values of $β_{0}$, $β_{1}$ and $P(Y_{i} = 1) = \pi_{i}$ are not observed, so we would not mention them and just use the estimates, $\hat{β}_{0}$, $\hat{β}_{1}$ and $\hat{\pi}_{i}$, from now on.

<font color='black'>

<br>

For each unit increase in *Body Size* ($x_{i}$), $logit(\hat{pi}_{i})$ increases by $0.7404$. In addition, if $x_{i} = 0$, then $logit(\hat{\pi}_{i}) = -21.9924$.

<br>

Now,
\begin{equation}
    \frac{\hat{\pi}_{i}}{1-\hat{\pi}_{i}} = e^{\hat{β}_{0} + \hat{β}_{1}x_{i}} = e^{\hat{β}_{0}}e^{\hat{β}_{1}x_{i}}
\end{equation}

Then, this leads to another interpretation, for each unit increase in *Body Size*, the odds of surviving increases multiplicatively by $e^{\hat{β}_{1}} = e^{0.7404} = 2.096774$.

$e^{\hat{β}_{1}}$ estimates the odds ratio of a success of survival for each additional one unit increase in $x_{i}$.
<br>
$e^{\hat{β}_{0}}$ estimates the odds of a success of survival when $x_{i}=0$.

Furthermore, $(e^{\hat{β}_{1}}−1) \cdot 100$% gives the percent increase in the odds of a success (surviving) for each one-unit increase in $x_{i}$.

We can do some further rearrangement of the formula and make $\hat{\pi}_{i}$, which is a probability, the subject of the formula. We will end up with:

\begin{equation}
    \hat{\pi}_{i} = \frac{ e^{\hat{β}_{0} + \hat{β}_{1}x_{i}} }{ 1 + e^{\hat{β}_{0} + \hat{β}_{1}x_{i}} }
\end{equation}

This is the formula that we use to predict probabilities and is used to plot the points in the logistic curve above. We should not use this logistic regression equation to predict the probability of surviving for any bodysize that is outside of the “experience of the model”. In other words, we cannot use this model to predict the survival probability for say a *Body Size* of $100$.

These are some other interpretations of the slope:
- $\hat{β}_{0} = 0$ implies that $P(Presence)$ is the same at each level of $x_{i}$.
- $\hat{β}_{0} > 0$ implies that $P(Presence)$ increases as $x_{i}$ increases.
- $\hat{β}_{0} < 0$ implies that $P(Presence)$ decreases as $x_{i}$ increases.

<br>
<br>

Now, we want to predict whether or not, $\hat{Y}_{i} = 0$ or $\hat{Y}_{i} = 1$. To do so, we use the predicted probabilities $\hat{\pi}_{i}$ and say that $\hat{Y}_{i} = 0$, if $\hat{\pi}_{i} ≤ threshold$ and $\hat{Y}_{i} = 1$, if $\hat{\pi}_{i} > threshold$. A common $threshold$ to use is $0.5$.

<br>

<font color='red'>
    
**Note:** The inequalities could be $<$ and $≥$ instead of $≤$ and $>$ respectively.

```python
# Plots for the Logistic Regression


# 1)
# Use the plt.scatter to get the scatter plot:
bodysizeplot = plt.scatter('Body Size', 'Survive', label = '$Y_{i}$', data=dat, color = 'magenta')


# 2)
# Using logistic regression to estimate the intercept and coefficient of Xbodysize:
bodysizelogitmod = sm.GLM(dat['Survive'], Xbodysize, family=sm.families.Binomial()).fit()

beta0 = bodysizelogitmod.params[0]
beta1 = bodysizelogitmod.params[1]


# Plot the logistic regression curve:
domain = np.linspace(start = 25, stop = 34, num = 10000)
plt.plot(domain.reshape((-1, 1)), expit(beta0 + domain * beta1).ravel(), color='green', label = '$Logit$ Curve', linewidth=2, markersize=1)

# expit is the sigmoid function defined as expit(x) = 1/(1+exp(-x)) = exp(x)/(exp(x)+1)
# The use of this was previously explained to get the probability, pihat_i.

# Plot the Fitted Values, Yhat_i, where we use the threshold of 0.5:
# So, Yhat_i = 1, if P(Yhat_i = 1|X) >= 0.5 and Yhat_i = 0, if P(Yhat_i = 1|X) < 0.5:
plt.scatter(Bodysize, (expit(beta0 + Bodysize * beta1).ravel() >= 0.5).astype(int), color = 'blue', marker = '+', label = "$\hat{Y}_{i}$", linewidth = 1, alpha = 1)


# Legend location:
plt.legend(loc='best', fontsize = 'large')

# Background colour of the plots:
sns.set_style('darkgrid')

# Size of the plot:
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(14, 7, forward = True)

# Define the title and axes labels and set the font sizes of them:
fig.suptitle('Logistic Regression Curve of Body Size vs Probability of Survival', fontsize=15)
plt.xlabel('Body Size', fontsize=12)
plt.ylabel('Survive', fontsize=12)


# function to show the plot:
plt.show()

#?plt.plot
#?expit
#?plt.scatter
```

```python
# Fitted Values of P(Yhat_i = 1|X):
ySurvive = bodysizelogitmod.predict(Xbodysize)

# Fitted Values of Yhat_i, where we use the threshold of 0.5.
# So, Yhat_i = 1, if P(Yhat_i = 1|X) >= 0.5 and Yhat_i = 0, if P(Yhat_i = 1|X) < 0.5:
ySurvive = (ySurvive >= 0.5).astype(int)


bodysizesummary = np.c_[ Survive, ySurvive, Bodysize ]
bodysizesummary = pd.DataFrame(bodysizesummary, columns=["Observed Survive", "Fitted Survive", "Body Size"])
bodysizesummary
```

### 5. Feature Selection



<br>

<font color='black'>

For both Regression and Classification projects, we will generally have to go through *feature selection* for datasets to determine the appropriate input variables to choose from.

We usually start off with a manual selection of variables from the dataset and remove those that do not impact the dependent variable based on the nature of the variable and/or prior knowledge in the field. Some form of this was conducted in the ETL Session, where the bitcoin data was transformed by removing certain columns that did not seem relevant. 


**Feature Selection:** Given a dataset, we identify which independent variables are appropriate for predicting the dependent variable. This can be done through:
<br>
1) Wrapper methods <br>
2) Filter methods <br>
3) Embedded methods <br>

<br>
<br>

Feature selection helps eliminate variables that are deemed *irrelevant* or *redundant*.
- **Irrelevant** in the sense that the variable would have no significant impact on the predicting the value of the dependent variable. For example, if we were to predict the price of a house, a variable such as the name of the construction company that built the house could be seen as irrelevant.

- **Redundant** in the sense that an independent variable may be a linear combination of the other independent variables. This means that the variable is linearly dependent. Recall in the Linear Algebra Session that a vector, **v<sub>1</sub>**, is linearly dependent to vectors, **v<sub>2</sub>** and **v<sub>3</sub>** if **v<sub>1</sub>** = a · **v<sub>2</sub>** + b · **v<sub>3</sub>**, where a and b are real numbers. We call this *perfect multicollinearity* and this is generally not good for modelling. A prime example where perfect multicollinearity is deterimental is in regression analysis where the parameters, $\hat{β}$, are estimated through least-squares method using a formula which contains the inverse of a transformation of the independent variables in matrix form (X<sup>T</sup>X)<sup>-1</sup>. This linear dependence is a problem because (X<sup>T</sup>X)<sup>-1</sup> would not exist. An example of redundant data in the case of pricing houses could be that one variable has the house measured in squared feet and another variable has the house measured in squared metres.

Feature selection can also help reduce **overfitting** in a model. Simply put, a model is said to overfit if it we have have selected the parameters and features of the model that fits the *training set of data* (sample data used to estimate model properties) with great accuracy but this model performs poorly on a *test set of data* (a different sample of the data). By reducing overfitting, the model would be more robust to better estimate new data. We look for a parsimonious model; one that does not have much predictor or independent variables, yet can generally estimate the dependent variable.

We generally try different models and see how well they perform against each other based on certain performance measures such as R<sup>2</sup>, adjusted R<sup>2</sup>, AIC, BIC or mean squared error.


<br>
<br>


In regression analysis, some of the basic methods we can use are an exhaustive search of all possible models and determine which one has the best performance measure score, or we could use automatic methods such as the greedy algorithms (examples include forward selection, backward elimination or bidirectional variants of each). The aforementioned menthods fall under wrapper methods.

The exhaustive search should provide the best model based on the performance criteria, but for datasets with many independent variables, this can be computationally expensive to run. The greedy algorithms run faster but may not necessarily find the best overall model.

The forward selection starts from the simplest model, either 0 or 1 independent variable(s) (or what the user predefines). It then works its way to the full model with all the variables (or what the user predefines) by adding one variable at a time. In each step of adding variables, the model with the best performance measure is kept and this value is stored. Even if adding a variable may reduce the score, we continue to add and kept track of the best model found in each step. In the end, the model with the best score out of all steps of the scores is the one selected. Backwards elimination is just the reverse, as you start off with the full model and eliminate variables to get to the simplest model; the best score and features in each step are also kept and evaluated at the end. Bidirectional variants of these involve the option of adding/eliminating variables (even if they were previously eliminated) in each step.


There are other methods which can be used using statistical software. Some of these can be done using the sklearn.feature_selection package in Python.
