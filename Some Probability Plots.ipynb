```python
#Import libraries

import numpy as np
from matplotlib import pyplot as plt # for plotting
from scipy.stats import norm # for generating a Normal distribution
from scipy.stats import t # for generating a t distribution
from scipy.stats import f # for generating an F distribution
from scipy.stats import chi2 # for generating a chi-square distribution
from scipy.stats import gamma # for generating a gamma distribution
from scipy.stats import binom # for generating a binomial distribution
from scipy.stats import poisson # for generating a poisson distribution
from scipy.stats import nbinom # for generating a binomial distribution
%matplotlib inline
```

```python
# Example of some Normal Distributions.

domain = np.linspace(start = -5, stop = 5, num = 1000)
# Return evenly spaced numbers over a specified interval.
# Returns `num` evenly spaced samples, calculated over the interval [`start`, `stop`].
# The endpoint of the interval can optionally be excluded.
# ?np.linspace


plt.plot(domain, norm.pdf(domain,loc=0,scale=np.sqrt(1)), label='μ=0, $σ^{2}$=1')
plt.plot(domain, norm.pdf(domain,loc=0,scale=np.sqrt(0.2)), label='μ=0, $σ^{2}$=0.2')
plt.plot(domain, norm.pdf(domain,loc=0,scale=np.sqrt(5)), label='μ=0, $σ^{2}$=5')
plt.plot(domain, norm.pdf(domain,loc=-2,scale=np.sqrt(0.5)), label='μ=-2, $σ^{2}$=0.5')
plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
# ?plt.plot
# ?norm.pdf
```

```python
# Example of some t Distributions.

domain = np.linspace(start = -5, stop = 5, num = 1000)
# Return evenly spaced numbers over a specified interval.
# Returns `num` evenly spaced samples, calculated over the interval [`start`, `stop`].
# The endpoint of the interval can optionally be excluded.
# ?np.linspace


plt.plot(domain, norm.pdf(domain,loc=0,scale=np.sqrt(1)), label='N(μ=0, $σ^{2}$=1)')
plt.plot(domain, t.pdf(domain,df=1,loc=0,scale=np.sqrt(1)), label='t(df=1)')
plt.plot(domain, t.pdf(domain,df=2,loc=0,scale=np.sqrt(1)), label='t(df=2)')
plt.plot(domain, t.pdf(domain,df=4,loc=0,scale=np.sqrt(1)), label='t(df=4)')
plt.plot(domain, t.pdf(domain,df=10,loc=0,scale=np.sqrt(1)), label='t(df=10)')
plt.title("Normal & t Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
# ?plt.plot
# ?norm.pdf
```

```python
# Example of some F Distributions.

domain = np.linspace(start = 0.01, stop = 5, num = 100000)
# Return evenly spaced numbers over a specified interval.
# Returns `num` evenly spaced samples, calculated over the interval [`start`, `stop`].
# The endpoint of the interval can optionally be excluded.
# ?np.linspace

plt.plot(domain, f.pdf(domain,dfn=1,dfd=1,loc=0,scale=1), label='df$_{1}$=1, df$_{2}$=1')
plt.plot(domain, f.pdf(domain,dfn=2,dfd=1,loc=0,scale=1), label='df$_{1}$=2, df$_{2}$=1')
plt.plot(domain, f.pdf(domain,dfn=3,dfd=2,loc=0,scale=1), label='df$_{1}$=5, df$_{2}$=2')
plt.plot(domain, f.pdf(domain,dfn=10,dfd=1,loc=0,scale=1), label='df$_{1}$=10, df$_{2}$=1')
plt.plot(domain, f.pdf(domain,dfn=100,dfd=100,loc=0,scale=1), label='df$_{1}$=100, df$_{2}$=100')
plt.title("F Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
# ?plt.plot
# ?f.pdf
```

```python
# Example of some chi-square Distributions.

domain = np.linspace(start = 0, stop = 8, num = 1000)
# Return evenly spaced numbers over a specified interval.
# Returns `num` evenly spaced samples, calculated over the interval [`start`, `stop`].
# The endpoint of the interval can optionally be excluded.
# ?np.linspace


plt.plot(domain, chi2.pdf(domain,df=1,loc=0,scale=np.sqrt(1)), label='df=1')
plt.plot(domain, chi2.pdf(domain,df=2,loc=0,scale=np.sqrt(1)), label='df=2')
plt.plot(domain, chi2.pdf(domain,df=3,loc=0,scale=np.sqrt(1)), label='df=3')
plt.plot(domain, chi2.pdf(domain,df=4,loc=0,scale=np.sqrt(1)), label='df=4')
plt.plot(domain, chi2.pdf(domain,df=6,loc=0,scale=np.sqrt(1)), label='df=6')
plt.plot(domain, chi2.pdf(domain,df=9,loc=0,scale=np.sqrt(1)), label='df=9')
plt.title("Chi-Square Distributions, $χ^{2}_{df}$")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.ylim((0,1))
plt.show()
# ?plt.plot
# ?chi2.pdf
```

```python
# Example of some Gamma Distributions.

domain = np.linspace(start = 0, stop = 20, num = 100000)
# Return evenly spaced numbers over a specified interval.
# Returns `num` evenly spaced samples, calculated over the interval [`start`, `stop`].
# The endpoint of the interval can optionally be excluded.
# ?np.linspace

plt.plot(domain, gamma.pdf(domain,a=1,loc=0,scale=2), label='$α$=1, $β$=0.5')
plt.plot(domain, gamma.pdf(domain,a=2,loc=0,scale=2), label='$α$=2, $β$=0.5')
plt.plot(domain, gamma.pdf(domain,a=3,loc=0,scale=2), label='$α$=3, $β$=0.5')
plt.plot(domain, gamma.pdf(domain,a=5,loc=0,scale=1), label='$α$=5, $β$=1')
plt.plot(domain, gamma.pdf(domain,a=9,loc=0,scale=0.5), label='$α$=9, $β$=2')
plt.plot(domain, gamma.pdf(domain,a=7.5,loc=0,scale=1), label='$α$=7.5, $β$=1')
plt.plot(domain, gamma.pdf(domain,a=0.5,loc=0,scale=1), label='$α$=0.5, $β$=1')
plt.title("Gamma Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.ylim((0,1))
plt.show()
# ?plt.plot
# ?gamma.pdf
```

```python
# Example of some Binomial Distributions.
fig, ax = plt.subplots(1, 1)

n = 20
p = 0.5
x = np.arange(binom.ppf(0.01, n, p), binom.ppf(0.99, n, p))
ax.plot(x, binom.pmf(x, n, p), 'bo', label='n=20, p=0.5')

n = 20
p = 0.7
x = np.arange(binom.ppf(0.01, n, p), binom.ppf(0.99, n, p))
ax.plot(x, binom.pmf(x, n, p), 'ro', label='n=20, p=0.7')

n = 40
p = 0.5
x = np.arange(binom.ppf(0.01, n, p), binom.ppf(0.99, n, p))
ax.plot(x, binom.pmf(x, n, p), 'mo', label='n=40, p=0.5')

plt.title("Binomial Distribution")
plt.xlabel("Value, $x$")
plt.ylabel("Probability, f($x$) = P($X=x$)")
plt.legend()
plt.show()
# ?plt.plot
# ?binom.pmf
```

```python
# Example of some Poisson Distributions.
fig, ax = plt.subplots(1, 1)

mu = 1
x = np.arange(poisson.ppf(0.01, mu), poisson.ppf(0.99, mu))
ax.plot(x, poisson.pmf(x, mu), 'bo', label='$λ$=1')

mu = 4
x = np.arange(poisson.ppf(0.01, mu), poisson.ppf(0.99, mu))
ax.plot(x, poisson.pmf(x, mu), 'ro', label='$λ$=4')

mu = 10
x = np.arange(poisson.ppf(0.01, mu), poisson.ppf(0.99, mu))
ax.plot(x, poisson.pmf(x, mu), 'mo', label='$λ$=10')

plt.title("Binomial Distribution")
plt.xlabel("Value, $x$")
plt.ylabel("Probability, f($x$) = P($X=x$)")
plt.legend()
plt.show()
# ?plt.plot
# ?poisson.pmf
```

```python
# Example of some Negative Binomial Distributions.
fig, ax = plt.subplots(1, 1)

n = 20
p = 0.5
x = np.arange(nbinom.ppf(0.01, n, p), nbinom.ppf(0.99, n, p))
ax.plot(x, nbinom.pmf(x, n, p), 'bo', label='r=20, p=0.5')

n = 20
p = 0.7
x = np.arange(nbinom.ppf(0.01, n, p), nbinom.ppf(0.99, n, p))
ax.plot(x, nbinom.pmf(x, n, p), 'ro', label='r=20, p=0.7')

n = 20
p = 0.3
x = np.arange(nbinom.ppf(0.01, n, p), nbinom.ppf(0.99, n, p))
ax.plot(x, nbinom.pmf(x, n, p), 'mo', label='r=20, p=0.3')

n = 40
p = 0.5
x = np.arange(nbinom.ppf(0.01, n, p), nbinom.ppf(0.99, n, p))
ax.plot(x, nbinom.pmf(x, n, p), 'go', label='r=40, p=0.5')



plt.title("Binomial Distribution")
plt.xlabel("Value, $k$")
plt.ylabel("Probability, f($k$) = P($X=k$)")
plt.legend()
plt.show()
# ?plt.plot
# ?nbinom.pmf
```
